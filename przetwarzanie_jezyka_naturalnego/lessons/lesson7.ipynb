```python
import os
import json
import pandas as pd
```

```python
pd.read_json('covid-19-papers/noncomm_use_subset/noncomm_use_subset/pmc_json/PMC1636417.xml.json')
```

```python
root_folder_path = 'covid-19-papers'

for root, dirnames, filenames in os.walk(root_folder_path):
    for filename in filenames:
        if filename != '.DS_Store':           
            with open(os.path.join(root, filename)) as file:
                paper_in_json = json.load(file)
                nycphil = pd.json_normalize(paper_in_json['body_text']) 
                print(nycphil.head())
```

```python
root_folder_path = 'covid-19-papers'

papers_dict = {}
for root, dirnames, filenames in os.walk(root_folder_path):
    for filename in filenames:
        if filename != '.DS_Store':           
            with open(os.path.join(root, filename)) as file:
                paper_in_json = json.load(file)
                papers_dict[paper_in_json['paper_id']] = paper_in_json['body_text']
```

```python
for key in papers_dict.keys():
    print(key)
```
